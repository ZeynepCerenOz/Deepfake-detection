```python
import kagglehub

# Download latest version of Celeb-DF v2
path = kagglehub.dataset_download("reubensuju/celeb-df-v2")

print("Path to dataset files:", path)
```

```python
import os

for root, dirs, files in os.walk(path):
    print(root)
    print("Folders:", dirs[:5])
    print("Files:", files[:5])
    print("-" * 50)
```

```python
import cv2
import matplotlib.pyplot as plt

video_path = "/kaggle/input/celeb-df-v2/Celeb-real/id1_0007.mp4"

cap = cv2.VideoCapture(video_path)

frames = []
count = 0

while True:
    ret, frame = cap.read()
    if not ret or count > 10:
        break

    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frames.append(frame)
    count += 1

cap.release()
```

```python
plt.figure(figsize=(15, 5))

for i in range(min(5, len(frames))):
    plt.subplot(1, 5, i+1)
    plt.imshow(frames[i])
    plt.axis("off")

plt.show()
```
